```python
import matplotlib.pyplot as plt
```

```python

# Vẽ biểu đồ: Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'c*-', label='Dữ liệu') # Vẽ đường 1
plt.plot(x1, y1, marker='s', color='k', linestyle='-', label='Dữ liệu-1')# Vẽ đường 2
plt.legend() # Hiển thị chú giải
plt.xlabel('Trục x') # Đặt nhãn cho trục x
plt.ylabel('Trục y') # Đặt nhãn cho trục y
plt.show() # Hiển thị biểu đồ
```

```python
# Vẽ biểu đồ cột Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [14, 11, 21, 19, 18]

plt.bar(categories, values, color='c', alpha=0.5) # Vẽ biểu đồ cột
plt.xlabel('Danh mục') # Đặt nhãn cho trục x
plt.ylabel('Giá trị') # Đặt nhãn cho trục y
plt.show() # Hiển thị biểu đồ
```

```python
# Vẽ biểu đồ hình tròn Pie Chart
my_labels = ['Ăn uống', 'Đi lại', 'Gia đình', 'Tiền nhà']
sizes = [25, 20, 15, 40]
my_colors = ['pink', 'gray', 'blue', 'orange']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ chi tiêu')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [21, 18, 11, 29, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán giông bão')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp trụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)  # Vẽ biểu đồ violin
plt.title('Biểu đồ violin ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Biểu đồ đám mây từ văn bản ví dụ')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, '*-', linewidth=2)
plt.fill(angles, values, 'c', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```

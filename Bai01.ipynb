```python
#Thêm thư viện
from PIL import Image
```

```python
# Khai báo một chứa thư mục (/ hoặc \\)
dir = 'H:/ThiGiacMayTinh/0101/img'
```

```python
# Khai báo đường dẫn ảnh
image_path = dir + "/2.JPG"
print(image_path)
```

```python
# Sử dụng Image.open() để đọc ảnh
img = Image.open(image_path)
```

```python
# Hiển thị ảnh
display(img)
```

```python
#In thông tin cơ bản
print(" Địn dạng ảnh: ", img.format)
```

```python
#In kích cở ảnh
print("Kích cỡ ảnh: ", img.size)
```

```python
# Lưu ảnh và chuyển đổi ảnh
new_img_2_path = dir + "/new_2.JPG"
new_img_PNG_2_path = dir + "/new_2.PNG"
img.save(new_img_2_path)
img.save(new_img_PNG_2_path)
```

```python
# Đóng tập tin
img.close()
```

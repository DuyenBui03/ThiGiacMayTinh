```python
#import thư viện
from PIL import Image
from imgtools import *
```

```python
my_path = 'H:/ThiGiacMayTinh/0101/img/5.JPG'
```

```python
#Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
#Định bị khu vực cắt
region = (130, 100, 190, 160)
#Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (100, 150)
img.paste(cropped_image, paste_position)
display(img)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'H:/ThiGiacMayTinh/0101/img/6.JPG'
# Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyển sanh ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=121)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên ảnh
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
#Đường dẫn ảnh
my_path = 'H:/ThiGiacMayTinh/0101/img/6.JPG'
# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")
#Hiển thị ảnh trên trục tọa độ
plt.imshow(im)
# Tạp một hình vẽ mới
plt.figure()
# Đặ màu sắc là xám
plt.gray()
# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, orgrin="image")
# Đảm bảo tỷ lệ trục x và trục y của biểu đồ là bằng nhau
plt.axis('equal')
# Hiển thị ảnh
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ng dn nh
my_path = 'H:/ThiGiacMayTinh/0101/img/6.JPG'
# c nh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyn sanh nh xm
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins=121)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Lm ni bt ng bin nh
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

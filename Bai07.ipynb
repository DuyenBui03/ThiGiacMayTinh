```python
# Vẽ đồ thị cơ bản
x = [1, 2, 3, 4, 5]
y = [11, 15, 25, 28, 28]
plt.plot(x, y, color='pink') # Màu sắc
plt.title('Ví dụ về biểu đồ') # Title
plt.show() # Hiển thị
```

```python
# Tắt hệ trục tọa độ
x = [1, 2, 3, 4, 5]
y = [11, 15, 25, 28, 28]

plt.plot(x, y, color='pink')  # Thay đổi màu thành đỏ
plt.title('Biểu đồ ví dụ với màu hồng')  # Thêm tiêu đề
# Tắt hệ trục tọa độ
plt.axis('off')  # Tắt hệ trục tọa độ
plt.show()
```
